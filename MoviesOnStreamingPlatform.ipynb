```python
# Linear model did not provide best fit while using Disney+ as predictors
# Use Polynomial instead

def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(1920, 1970, 2000)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Rotten Tomatoes Ratings')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Rotten Tomatoes Ratings')

    plt.show()
    plt.close()
```

Decision making : determining a good model fit

When comparing models, the model with higher R-squared value and the model with the smallest value is a better fit for the data.

1.Simple Linear Regression: using Year as a predictor variable of Rotten Tomatoes
R-squared: 0.0012675183282803149
MSE: 173.8116948329111

2.Multiple Linear Regression: using Prime Video, Disney+, Netflix, Year
R-squared: 0.05448215270340562
MSE: 164.55063047344186

3.Polynomial Fit: using Year as a predictor variable of Rotten Tomatoes
R-squared: 0.02060100504058482
MSE:170.44704398379636

__Simple Linear Regression Model (SLR) vs Multiple Linear Regression Model (MLR)__

R-Squared: In this case, we can also see that there is a big difference between the R-squared of the SLR and the R-squared of the MLR. The R-squared for the SLR (0.0012675183282803149) is very small compared to the R-squared for the MLR (0.05448215270340562).
MSE: The MSE of MLR is 164.55063047344186 while SLR has an 173.8116948329111. The MSE of MLR is much smaller.

This R-squared in combination with the MSE show that MLR seems like the better model fit in this case compared to SLR.

__Simple Linear Model (SLR) vs. Polynomial Fit__

MSE: We can see that Polynomial Fit brought down the MSE, since this MSE is smaller than the one from the SLR.
R-squared: The R-squared for the Polynomial Fit is larger than the R-squared for the SLR, so the Polynomial Fit also brought up the R-squared quite a bit.

Since the Polynomial Fit resulted in a lower MSE and a higher R-squared, we can conclude that this was a better fit model than the simple linear regression for predicting "Rotten Tomatoes" with "Year" as a predictor variable.

__Multiple Linear Regression (MLR) vs. Polynomial Fit__

MSE: The MSE for the MLR is smaller than the MSE for the Polynomial Fit.
R-squared: The R-squared for the MLR is also much larger than for the Polynomial Fit.

Comparing these three models, we conclude that the MLR model is the best model to be able to predict price from our dataset.

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width =12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.histplot(RedFunction, color="r", label=RedName, bins=20)
    ax2 = sns.histplot(BlueFunction, color="b", label=BlueName, bins=20)

    plt.title(Title)
    plt.xlabel('Rotten Tomatoes Ratings')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    plt.close()
```

```python
def PollyPlot(xtrain, xtest, y_train, y_test, lr, poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    # Scatter plot for training data
    plt.scatter(xtrain, y_train, color='red', label='Training Data')

    # Scatter plot for test data
    plt.scatter(xtest, y_test, color='green', label='Test Data')

    # Generate x values for plotting
    x_values = np.linspace(min(xtrain.min(), xtest.min()), max(xtrain.max(), xtest.max()), 100)

    # Transform x_values using polynomial transformation
    x_values_poly = poly_transform.fit_transform(x_values.reshape(-1, 1))

    # Predicted values using linear regression
    y_values = lr.predict(x_values_poly)

    # Plot the predicted function
    plt.plot(x_values, y_values, label='Predicted Function', color='blue')

    plt.ylim([0, 100])
    plt.xlabel('Prime Video')
    plt.ylabel('Rotten Tomatoes')
    plt.legend()
    plt.show()
```

```python
# Ridge Regression to see how the parameter alpha changes the model. Test data will be used as validation data 

pr =  PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train[["Prime Video", "Disney+","Netflix", "Year", "Age"]])
x_test_pr = pr.fit_transform(x_test[["Prime Video", "Disney+","Netflix", "Year", "Age"]])
```

```python
from sklearn.linear_model import Ridge
```

```python
RidgeModel = Ridge(alpha=0.1)
```

```python
RidgeModel.fit(x_train_pr, y_train)
```

```python
yhat = RidgeModel.predict(x_test_pr)
```

```python
# compare the first five predicted samples to our test set:
print('predicted:', yhat[0:4])
print('test set:', y_test[0:4].values)
```

```python
# select the value of alpha that minimizes the test error. Also, created a progress bar to see how many iterations we have completed.

from tqdm import tqdm

Rsqu_test = []
Rsqu_train =  []
dummy1 = []
Alpha =  10 * np.array(range(0, 1000))
pbar =  tqdm(Alpha)

for alpha in pbar:
    RidgeModel = Ridge(alpha=alpha)
    RidgeModel.fit(x_train_pr, y_train)
    test_score, train_score = RidgeModel.score(x_test_pr, y_test), RidgeModel.score(x_train_pr, y_train)

    pbar.set_postfix({"Test Score" : test_score, "Train Score" : train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
width = 12
height =10
plt.figure(figsize=(width, height))

plt.plot(Alpha, Rsqu_test, label='validation data ')
plt.plot(Alpha, Rsqu_train, 'r', label='training data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```
